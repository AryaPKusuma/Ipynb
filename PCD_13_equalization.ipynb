```python
import cv2
import matplotlib.pyplot as plt

def equalization(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    equalized_image = cv2.equalizeHist(gray)
    return equalized_image

def analize_histogram(image):
    hist = cv2.calcHist([image], [0], None, [256], [0, 256])
    return hist

def show_histogram(image, title):
    plt.plot(image)
    plt.title(title)
    plt.xlabel('Intensitas Piksel')
    plt.ylabel('Frekuensi')
    plt.show()

def show_image(image, text):
    plt.imshow(image, cmap='gray')
    plt.title(text)
    plt.axis('off')
    plt.show()

image = cv2.imread('image.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

equalized_image = equalization(image)
image_before_equalization = analize_histogram(image)
image_after_equalization = analize_histogram(equalized_image)

show_image(gray, "before equalization")
show_histogram(image_before_equalization, "before equalization")
show_image(equalized_image, "after equalization")
show_histogram(image_after_equalization, "after equalization")
```
