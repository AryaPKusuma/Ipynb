```python
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import pandas as pd
```

```python
## load kata slang ke dalam sebuah dataframe dan memberikan kolom slang dan baku pada dataframenya
slang_word = pd.read_csv('/content/id-multi-label-hate-speech-and-abusive-language-detection/new_kamusalay.csv',
                         encoding='latin-1', names=['slang', 'baku'])
slang_dict = dict(zip(slang_word['slang'], slang_word['baku']))
```

```python
nltk.download('punkt')
nltk.download('stopwords')
factory = StemmerFactory()
stemmer = factory.create_stemmer()

## fungsi untuk preproses kata
def preprocess_text(text):
  text = text.lower()
  text = re.sub(r'\s+', ' ', text)
  text = re.sub(r'http\S+', '', text)
  text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
  tokens = nltk.word_tokenize(text)
  stemmed_tokens = [stemmer.stem(token) for token in tokens]
  filtered_tokens = [token for token in stemmed_tokens if token not in string.punctuation]
  text = ' '.join(stemmed_tokens)
  return text

## fungi untuk memfilter kata slang pada komentar
def filter_slang(text, slang_dict):
    words = text.split()
    replaced_text = ' '.join([slang_dict.get(word, word) for word in words])
    return replaced_text

## fungsi untuk mengubah nilai sentimen dari kategori angka ke kata
def categorize_sentiment(sentiment):
  if sentiment == 0:
    return "Netral"
  elif sentiment == 1:
    return "Positif"
  elif sentiment == 2:
    return "Negatif"
  else:
    raise ValueError("Invalid sentiment value")
```

```python
# load data berlabel ke dataframe
comments = pd.read_csv('/content/labeled_data.csv')
comments.head()
```

```python
## nengecek nilai sentimen
unique_values = comments['sentiment'].unique()
print(unique_values)
```

```python
## mengubah nilai sentimen yang sebelumnya angka menjadi kategori positif negatif netral
comments = comments[comments['sentiment'].isin([0, 1, 2])]
comments.loc[:, 'sentiment'] = comments['sentiment'].astype(int)
comments.loc[:, 'sentiment'] = comments['sentiment'].apply(categorize_sentiment)
```

```python
comments.head()
```

```python
comments['text_processed'] = comments['text'].apply(filter_slang, args=(slang_dict,))
comments['text_processed'] = comments['text_processed'].apply(preprocess_text)
```

```python
# Hapus baris dengan nilai NaN pada kolom 'text_processing'
comments['text_processed'].replace('', pd.NA, inplace=True)
comments.dropna(subset=['text_processed'], inplace=True)
```

```python
comments.isnull().sum()
```

```python
comments.to_csv('comments_processed.csv', index=False)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from imblearn.over_sampling import SMOTE
```

### Membuat data Traing

```python
# split data test dan data training
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
# menerapkan SMOTE
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)
```

### Model

```python
#model tanpa smote
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)

```

### Model dengan SMOTE

```python
# Model dengan SMOTE
clf_smote = DecisionTreeClassifier(random_state=42)
clf_smote.fit(X_train_res, y_train_res)
```

## Penilaian peforma model

```python
## menilai performa menggunakan cross validation
from sklearn.model_selection import cross_val_score

cv_scores = cross_val_score(clf, X, y, cv=5)

print("Cross-validation Scores:", cv_scores)
print("Mean Accuracy:", np.mean(cv_scores))
```

```python
# score akurasi menggunakan accuracy_score

y_pred = clf.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

## menilai performa model+smote

```python
## menilai performa dengan smote
cv_scores_smote = cross_val_score(clf_smote, X, y, cv=5)

print("Cross-validation Scores:", cv_scores_smote)
print("Mean Accuracy:", np.mean(cv_scores_smote))
```

```python
y_pred_smote = clf_smote.predict(X_test)
print("Accuracy (with SMOTE):", accuracy_score(y_test, y_pred_smote))
print("Classification Report (with SMOTE):")
print(classification_report(y_test, y_pred_smote))
```

## Hyperparameter

```python
# Definisikan grid parameter yang akan dicari
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid_search = GridSearchCV(estimator=clf_smote, param_grid=param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train_res, y_train_res)
print("Best Parameters:", grid_search.best_params_)

best_model = grid_search.best_estimator_
```

```python
y_pred_best = best_model.predict(X_test)
print("Accuracy (Best Model):", accuracy_score(y_test, y_pred_best))
print("Classification Report (Best Model):")
print(classification_report(y_test, y_pred_best))
```

```python
cv_scores_best = cross_val_score(best_model, X, y, cv=5)
print("Cross-validation Scores (Best Model):", cv_scores_best)
print("Mean Accuracy (Best Model):", np.mean(cv_scores_best))
```
