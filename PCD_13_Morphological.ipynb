```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def dilation(image):
    kernel = np.ones((5, 5), np.uint8)
    dilated_image = cv2.dilate(image, kernel, iterations=1)

    plt.subplot(1, 3, 2)
    plt.imshow(dilated_image, cmap='gray')
    plt.title('dilation Image')
    plt.axis('off')
    plt.show()

def erosion(image):
    kernel = np.ones((5, 5), np.uint8)
    erosion_image = cv2.erode(image, kernel, iterations=1)

    plt.subplot(1, 3, 2)
    plt.imshow(erosion_image, cmap='gray')
    plt.title('erotion Image')
    plt.axis('off')
    plt.show()

def opening(image):
    kernel = np.ones((5, 5), np.uint8)
    img_opening = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)

    plt.subplot(1, 3, 2)
    plt.imshow(img_opening, cmap='gray')
    plt.title('opening Image')
    plt.axis('off')
    plt.show()

def closing(image):
    kernel = np.ones((5, 5), np.uint8)
    img_closing = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)

    plt.subplot(1, 3, 2)
    plt.imshow(img_closing, cmap='gray')
    plt.title('closing Image')
    plt.axis('off')
    plt.show()

image = cv2.imread('hasil_masking.png')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
_, binary_image = cv2.threshold(gray, 128, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(binary_image, cmap='gray')
plt.title('before dilation Image')
plt.axis('off')
dilation(binary_image)
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(binary_image, cmap='gray')
plt.title('normal Image')
plt.axis('off')
erosion(binary_image)
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(binary_image, cmap='gray')
plt.title('normal Image')
plt.axis('off')
opening(binary_image)
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(binary_image, cmap='gray')
plt.title('normal Image')
plt.axis('off')
closing(binary_image)
```
